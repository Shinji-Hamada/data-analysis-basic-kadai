```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3,tight_layout=True)

y = [1, 2, -5, 2]
plt.plot(y)
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

x = np.linspace(0, 10, 100)
x0 = 2 + 2 * np.sin(2 * x)
plt.plot(x)
axes[0,1].plot(x0)

w = np.random.normal(15, 5, 2000)
plt.hist(w, bins=30)
axes[0,2].hist(w, bins=30)

np.random.seed(3)
z = 4 + np.random.normal(0, 2, 60)
z0 = 4 + np.random.normal(0, 2, len(z))
plt.scatter(z, z0)
axes[1,0].scatter(z, z0)
```
